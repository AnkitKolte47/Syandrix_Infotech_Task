### 1st task calculating running total using windows function


```python
my_window = Window.partitionBy("category")\
.orderBy("order_date")\
.rowsBetween(Window.unboundedPreceding,Window.currentRow)
```

```python
running_total = orders_df.withColumn("running_total",round(sum('final_amount').over(my_window)))
```

```python
display(running_total)
```

### 2nd Task diffrence previous values

```python
My_window = Window.partitionBy("category")\
.orderBy("order_date")
```

```python
previous_value = orders_df.withColumn("previous_value",lag("final_amount").over(My_window))
```

```python
display(previous_value)
```

```python
values_diff = previous_value.withColumn("values_diffrence",expr("final_amount - previous_value"))
```

```python
display(values_diff)
```
